```python
import pandas as pd
import numpy as np
```

```python
try:
   df=pd.read_csv("AQI_data.csv")  #READING THE CSV FILE #
   print(df)

except:
     print("file not present")
```

```python
#Show the dtype and number of  values in each column #
print("datatypes of each columns are \n ", df.dtypes)
```

```python
# SHOW ALL non-null values in each column #
print("non-null values are")  
df.notnull().sum()      #PRINTING ALL THE NON-NULL VALUES #
```

```python
# Use NUMPY to compute the mean AQI, max PM2.5, and min PM10 values for each city
city_stats = df.groupby('City').agg({
    'AQI': lambda x: np.mean(x),      #using lambda function #
    'PM2.5': lambda x: np.max(x),
    'PM10': lambda x: np.min(x)
}).reset_index()

print(city_stats)
```

```python
# Group the dataset by City and calculate the required statistics
city_stats = df.groupby('City').agg({
    'AQI': 'mean',
    'PM2.5': 'max',
    'PM10': 'min'
}).reset_index()

# Converting the DataFrame to a dictionary
result_dict = city_stats.set_index('City').T.to_dict('list')

print("AQI, maximum PM2.5, and minimum PM10 values for each city:")
print(result_dict)
```
